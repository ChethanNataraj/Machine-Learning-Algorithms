# Logistic Regression 

```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
%matplotlib inline

cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify = cancer.target, random_state = 42)

log_reg = LogisticRegression()

log_reg.fit(X_train, y_train)
```

```python
print('Accuracy of the training set: {:.3f}'.format(log_reg.score(X_train, y_train)))
print('Accuracy of the test set: {:.3f}'.format(log_reg.score(X_test, y_test)))
```

### Let's try to adjust the default parameters to see how the algorithm performs

Regularization:

c = lower c adjusts to the majority of datapoints
higher c: correctly classify each data point

```python
log_reg100 = LogisticRegression(C = 100)
log_reg100.fit(X_train, y_train)
print('Accuracy of the training set: {:.3f}'.format(log_reg100.score(X_train, y_train)))
print('Accuracy of the test set: {:.3f}'.format(log_reg100.score(X_test, y_test)))
```

```python
log_reg001 = LogisticRegression(C = 0.01)
log_reg001.fit(X_train, y_train)
print('Accuracy of the training set: {:.3f}'.format(log_reg001.score(X_train, y_train)))
print('Accuracy of the test set: {:.3f}'.format(log_reg001.score(X_test, y_test)))
```

```python
plt.plot(log_reg.coef_.T, 'o', label='C=1')
plt.plot(log_reg100.coef_.T, '^', label='C=100')
plt.plot(log_reg001.coef_.T, 'v', label='C=0.01')
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation = 90)
plt.hlines(0,0,cancer.data.shape[1])
plt.ylim(-5,5)
plt.xlabel('Co-efficient index')
plt.ylabel('Co-efficient magnitude')
plt.legend()
```
